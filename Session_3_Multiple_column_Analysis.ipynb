```python
for i in cat[1:]:    # i is col name
    keys=visa_df[i].value_counts().keys()  # i is col name
    values=visa_df[i].value_counts().values # i is col name
    plt.figure(figsize=(10,4))
    plt.subplot(3,3,i)    # i is number
    plt.bar(keys,values)
```

```python
plt.figure(figsize=(16,8))
for i in range(1,len(cat)):    # i is col name
    keys=visa_df[cat[i]].value_counts().keys()  # i is cat[i] name
    values=visa_df[cat[i]].value_counts().values # i is cat[i] name
    plt.subplot(3,3,i)    # i is number
    plt.bar(keys,values)  
```

```python
plt.figure(figsize=(22,10))
for i in range(1,len(cat)):    # i is col name
    keys=visa_df[cat[i]].value_counts().keys()  # i is cat[i] name
    values=visa_df[cat[i]].value_counts().values # i is cat[i] name
    plt.subplot(3,3,i).bar(keys,values)
    plt.title(f'{cat[i]}_bar_chart')
    plt.suptitle('Barcharts')
```

**pie chart**

```python
plt.figure(figsize=(22,10))
for i in range(1,len(cat)):
    keys=visa_df[cat[i]].value_counts().keys()
    values=visa_df[cat[i]].value_counts().values
    plt.subplot(3,3,i).pie(values,
                           labels=keys,
                           radius=1,
                           autopct='%0.2f%%',
                           startangle=45)
    plt.title(f'{cat[i]}_pie_chart')
    plt.suptitle('Pie Chart')
```

**Z score**

  

```python
pwage_df=visa_df['prevailing_wage']
pwage_men=np.mean(pwage_df)
pwage_std=np.std(pwage_df)
```

```python
dif=pwage_df-pwage_men
```

```python
visa_df['prevailing_wage_z']=dif/pwage_std
```

```python
csv_path=r"C:\Users\subra\Documents\NareshIT\datafiles\Visadataset.csv"
visa_df=pd.read_csv(csv_path)
cat=visa_df.select_dtypes(include='object').columns
num=visa_df.select_dtypes(exclude='object').columns
```

```python
pwage_df=visa_df['prevailing_wage']
min=pwage_df.min()
max=pwage_df.max()

up=pwage_df-min
lp=(max-min)
visa_df['prevailing_wage_N']=up/lp
visa_df
```

```python
from sklearn.preprocessing import MinMaxScaler
mm =MinMaxScaler()

visa_df['prevailing_wage_N']=mm.fit_transform(visa_df[['prevailing_wage']])
```

```python
visa_df[['prevailing_wage','prevailing_wage_N']]
```
